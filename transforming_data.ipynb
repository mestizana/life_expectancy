```python
# 1. Read in data:
df2 = pd.read_csv('../data/gapminder_total.csv', index_col=0)
df2
```

```python
df2.info()
```

```python
# 2. What is the median population in the data set?
df2['population'].median()
```

```python
# 3. How often does each continent appear in the data set?
df2['continent'].value_counts()
```

```python
# 4. Which continent has the lowest average fertility rate overall?
df2.groupby('continent')['fertility'].mean()
```

```python
# 5. What was the average life expectancy in Europe in 2015? 
# Hint: first filter for 2015 then apply groupby.

lf_2015 = df2[df2['year']==2015]
lf_2015
```

```python
lf_2015.groupby('continent')['life expectancy'].mean()
```

```python
# 6. How many countries does each continent have in the dataset?
# Hint: filter for one year and count

lf_2000 = df2[df2['year']==2000]
lf_2000
```

```python
lf_2000['continent'].value_counts()
```

```python
# 7. What is the average population of a European country in 1976 compare to 2015?
# Hint: once again filter for the year in question and do each year separately to compare

lf_1976 = df2[df2['year']==1976]
lf_1976
```

```python
lf_2015 = df2[df2['year']==2015]
lf_2015
```

```python
lf_1976.groupby('continent')['population'].mean()
```

```python
lf_2015.groupby('continent')['population'].mean()
```

```python
# 8. What is the highest population a continent ever had?
# Hint: group by multiple columns

one = df2.groupby(['continent', 'year'])['population'].max()
one
```

```python
two = one.groupby('continent').max()
two
```

```python
# 9. Which continent had that population and in which year?
# Hint: group by multiple columns and filter for the result from #6

three = one.idxmax()
three
```

```python
# 10. Plot a bar plot comparison of life_expectancy, fertility, population average per continent
# Hint: for a quick and dirty graph use Pandas ploting ability
# Hint: use log=True to make the visual more friendly to the eyes

```

```python

life_expectancy_mean = df2.groupby('continent')['life expectancy'].mean()
fertility_mean = df2.groupby('continent')['fertility'].mean()
population_mean = df2.groupby('continent')['population'].mean()

cont = df2['continent']
my_list = [life_expectancy_mean , fertility_mean, population_mean]


plt.figure(figsize=(9,5))
plt.bar(x = cont ,
        y = my_list , 
        height = 0.5,
        log=True)

plt.legend()
plt.xlabel('continent')
plt.ylabel('average')
plt.title("comparison of life expectancy, fertility and population average per continent");
```
